```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
    
    # Visit https://galaxyfacts-mars.com/
url = "https://galaxyfacts-mars.com/"
browser.visit(url)

time.sleep(1)

    # Scrape page into Soup
html = browser.html
soup = bs(html, "html.parser")

# Get the facts about the planet including Diameter, Mass, etc.
mars_facts = soup.find_all("ul", class_="mt-3")
for facts in mars_facts:
    print(facts)
    
```

```python
mars_profile = soup.find_all("div", class_="sidebar")
for facts in mars_profile:
    print(facts)
```
